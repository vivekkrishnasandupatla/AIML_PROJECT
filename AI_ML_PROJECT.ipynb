```python

x = d.iloc[:, :-1].values
y = d.iloc[:, -1].values
print(x.shape,y.shape)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
print(x_train)
print(y_train)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix,accuracy_score,r2_score,classification_report,mean_squared_error
model1=LinearRegression()
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
##pt.scatter(x_train,y_train)
#pt.plot(x_test, model.predict(x_test), color = 'blue')
#pt.show()
```

```python
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

KNN

```python
from sklearn.neighbors import KNeighborsRegressor
model2=KNeighborsRegressor()
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
print(y_pred)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

SVM

```python
from sklearn.svm import SVR
model3=SVR()
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
print(y_pred)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model4=DecisionTreeRegressor()
model4.fit(x_train,y_train)
y_pred=model4.predict(x_test)
print(y_pred)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
from sklearn import tree
tree.plot_tree(model4,filled=True)
```
